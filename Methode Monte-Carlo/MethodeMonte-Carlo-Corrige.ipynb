# Calcul approché d'une intégrale par la méthode de Monte-Carlo
--- 

## Objectif 
L'objectif de ce TP est de voir une nouvelle méthode pour calculer une valeur approchée d'une intégrale, lorsque la méthode des rectangles est difficilement utilisable.
    
L'intégrale à calculer est la suivante: $ \displaystyle I = \int_0^1 f(x) dx$\, avec $f(x) = x \, (1-x) \, \sin^2\big(200 \, x \, (1-x)\big)$ 
    
La courbe représentative de la fonction $f$ sur $[0 ; 1]$ est la suivante:
    
![Fig1](https://raw.githubusercontent.com/HeleneLanglais/TP-Python/master/Methode%20Monte-Carlo/Images/Fig1.png)
    
Les variations de la courbe ne permettent pas d'appliquer facilement la méthode des rectangles pour calculer une valeur approchée de l'intégrale $I$ et il est très compliqué de déterminer une primitive de $f$.

## Méthode de Monte-Carlo:
Cette méthode va nous permettre d'estimer une aire et donc de déterminer une valeur approchée de l'intégrale $I$.

On commence par choisir un rectangle qui contient le domaine $D$ dont on cherche l'aire.

Puis on choisit au hasard $n$ points dans ce rectangle.

On compte ensuite, parmi les $n$ points, ceux qui sont postionnés dans le domaine $D$. (Notons $n_D$ ce nombre de points)

Si le nombre de points $n$ est assez grand, et si les points ont bien été choisis au hasard, on aura alors $\dfrac{n_D}{n} \approx \dfrac{Aire(D)}{Aire(Rectangle)}$

On obtient ainsi une valeur approchée de $I$ : $I = Aire(D) \approx \dfrac{n_D}{n} \times {Aire(Rectangle)}$

![Fig2](https://raw.githubusercontent.com/HeleneLanglais/TP-Python/master/Methode%20Monte-Carlo/Images/Fig2.png)

## Ecriture du programme:
 Pour pouvoir utiliser les fonctions mathématiques et de tirage au sort, nous allons avoir besoin de deux bibliothèques

```python
import math as m
import random as rd
```

Initialisation des variables: A vous de compléter les 4 premières lignes du code en vous aidant du graphique précédent.

```python
xmin = 0                      # abscisse minimale du rectangle
xmax = 1                      # abscisse maximale du rectangle
ymin = 0                      # ordonées minimale du rectangle
ymax = 0.25                   # ordonées maximale du rectangle

n = 1000                      # Nombre de points tirés au hasard
nD = 0                        # Nombre de points dans le domaine, initialisé à 0 
```

Création d'une fonction pour calculer image par la fonction $f$:

```python
def image(x):
    y = x*(1-x)*(m.sin(200*x*(1-x)))**2        # calcul de l'image de x par la fonction f en utilisant la fonction sinus via la commande m.sin
    return(y)                                  # y est le résultat retourné par la fonction "image"
```

Il faut ensuite créer tous les points, puis regarder pour chacun d'entre eux, s'il est dans le domaine ou bien à l'extérieur.

Dans notre cas, il faut que le point soit sous la courbe pour être dans le domaine.

Compléter le code suivant pour atteindre cet objectif, en utilisant la fonction image écrite précédemment.

```python
for i in range (0,n):
    x = rd.uniform(xmin, xmax)          # tirage aléatoire (loi uniforme) de l'abscisse du point
    y = rd.uniform(ymin, ymax)          # tirage aléatoire (loi uniforme) de l'ordonnéee du point
    if y <= image(x):
        nD = nD+1
```

La dernière étape consiste à calculer une valeur approchée de $I$ avec le résultat vu précédemment: $I = Aire(D) \approx \dfrac{n_D}{n} \times {Aire(Rectangle)}$

Compléter le code

```python
I = (nD/n)*(xmax-xmin)*(ymax-ymin)        
print("\033[34m Une valeur approchée de I est:",I,"\033[0m ")
```

## Programme complet

```python
import math as m
import random as rd

xmin = 0                      # abscisse minimale du rectangle
xmax = 1                      # abscisse maximale du rectangle
ymin = 0                      # ordonées minimale du rectangle
ymax = 0.25                   # ordonées maximale du rectangle

n = 10000                       # Nombre de points tirés au hasard
nD = 0                        # Nombre de points dans le domaine, initialisé à 0

def image(x):
    y = x*(1-x)*(m.sin(200*x*(1-x)))**2        # calcul de l'image de x par la fonction f
    return(y)
      
for i in range (0,n):
    x = rd.uniform(xmin, xmax)          # tirage de l'abscisse du point
    y = rd.uniform(ymin, ymax)          # tirage de l'ordonnéee du point
    if y <= image(x):
        nD = nD+1

I = (nD/n)*(xmax-xmin)*(ymax-ymin)        
print("\033[34m Une valeur approchée de I est:",I,"\033[0m ")  
```

## Complément n°2: Fluctuation de la valeur approchée de $I$
Nous allons reprendre le programme initial et l'utiliser pour étudier les fluctuations de la valeur aprochée de $I$ obtenue par la méthode de Monte-Carlo.

Pour cela nous allons faire tourner notre algorithme plusieurs fois de suite.

Il faut donc créer une boucle `for` qui va contenir le programme précédent.    

```python
import math as m
import random as rd
import matplotlib.pyplot as plt  

xmin = 0                      # abscisse minimale du rectangle
xmax = 1                      # abscisse maximale du rectangle
ymin = 0                      # ordonées minimale du rectangle
ymax = 0.25                   # ordonées maximale du rectangle

n = 1000                      # Nombre de points tirés au hasard

k = 10                        # Nombre d'itération de la méthode
ListeI = []                   # Liste qui va contenir les différentes valeurs approchées de I

def image(x):
    y = x*(1-x)*(m.sin(200*x*(1-x)))**2        # calcul de l'image de x par la fonction f
    return(y)

for j in range(0,k):                        # boucle pour répétée k fois un calcul de I
    nD = 0                                  # Nombre de points dans le domaine, initialisé à 0
    for i in range (0,n):
        x = rd.uniform(xmin, xmax)          # tirage de l'abscisse du point
        y = rd.uniform(ymin, ymax)          # tirage de l'ordonnéee du point
        if y <= image(x):
            nD = nD+1
    
    I = (nD/n)*(xmax-xmin)*(ymax-ymin)
    ListeI.append(I)                        # La fonction .append permet d'ajouter un élément (ici I) à la liste "ListeI" 
print( "\033[34m La liste des valeurs de I est: ",ListeI,"\033[0m")   
```

Pour visualiser ces fluctuations, nous pouvons créer un nouveau graphique, avec en abscisse le numéro de l'essai et en ordonnée la valeur approchée de $I$ obtenue.

```python
NumeroEssai = [i for i in range(k)]                # permet de créer une liste [ 0,1,2,...], qui sera la liste des abscisses
plt.figure(figsize=(15,6))                         # Choix de la taille du graphique
plt.ylim([0.06,0.10])                              # affiche l'ordonnée entre 0.06 et 0.10
plt.scatter(NumeroEssai,ListeI, s=5)               # création du graphique, s=2 prmet de régler la taille des points
plt.show()                                         # affichage du graphique
```

Pour cela, nous allons définir un domaine: le demi-cercle supérieur, centré en $O$ et de rayon 1.

![Fig3](https://raw.githubusercontent.com/HeleneLanglais/TP-Python/master/Methode%20Monte-Carlo/Images/Fig3.png)

```python
 import math as m
import random as rd

xmin = -1                     # abscisse minimale du rectangle
xmax = 1                      # abscisse maximale du rectangle
ymin = 0                      # ordonées minimale du rectangle
ymax = 1                      # ordonées maximale du rectangle

n = 100000                     # Nombre de points tirés au hasard
nD = 0                        # Nombre de points dans le domaine, initialisé à 0

def image(x):
    y = m.sqrt(1-x**2)        # calcul de l'image de x par la fonction f
    return(y)
      
for i in range (0,n):
    x = rd.uniform(xmin, xmax)          # tirage de l'abscisse du point
    y = rd.uniform(ymin, ymax)          # tirage de l'ordonnéee du point
    if y <= image(x):
        nD = nD+1

I = (nD/n)*(xmax-xmin)*(ymax-ymin)        
print("\033[34m Une valeur approchée de l'aire du demi-cercle est:",I,"\033[0m ")  
print("\033[34m Une valeur approchée de pi est:",2*I,"\033[0m ")                          # Aire 1/2 cercle = pi * R^2 /2 = pi/2 , Donc pi = 2 * Aire 1/2 cercle
```
