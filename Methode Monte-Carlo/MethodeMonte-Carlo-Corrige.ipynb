## <font color='steelblue'> Objectif 
<font color='steelblue'>L'objectif de ce TP est de voir une nouvelle méthode pour calculer une valeur approchée d'une intégrale, lorsque la méthode des rectangles est difficiliment utilisable.
    
<font color='steelblue'>L'intégrale à calculer est la suivante: $ \displaystyle I = \int_0^1 f(x) dx$\, avec $f(x) = x \, (1−x) \, \sin^2\big(200 \, x \, (1−x)\big)$ 
    
<font color='steelblue'> La courbe représentative de la fonction $f$ sur $[0 ; 1]$ est la suivante:
    
<img src ="Fig1.png" style="width:800px">
    
<font color='steelblue'> Les variations de la courbe ne permettent pas d'appliquer facilement la méthode des rectangles pour calculer une valaur approchée de l'intégrale $I$ et il est très compliqué de déterminer une primitive de $f$.

## <font color='green'> Méthode de Monte-Carlo:
<font color='green'> Cette méthode va nous permettre d'estimer une aire et donc de déterminer une valeur approchée de l'intégrale $I$.\
<font color='green'> On commence par choisir un rectangle qui contient le domaine $D$ dont on cherche l'aire.\
<font color='green'> Puis on choisit au hasard $n$ points dans ce rectangle.\
<font color='green'> On compte ensuite, parmi les $n$ points, ceux qui sont postionnés dans le domaine $D$. (Notons $n_D$ ce nombre de points)\
<font color='green'> Si le nombre de points $n$ est assez grand, et si les points ont bien été choisis au hasard, on aura alors $\dfrac{n_D}{n} \approx \dfrac{Aire(D)}{Aire(Rectangle)}$

<img src ="Fig2.png" style="width:800px">

<font color='green'> On obtient ainsi une valeur approchée de $I$ : $I = Aire(D) \approx \dfrac{n_D}{n} \times {Aire(Rectangle)}$


## <font color='steelblue'> Ecriture du programme:
<font color='steelblue'>Pour pouvoir utiliser les fonctions mathématiques et de tirage au sort, nous allons avoir besoin de deux bibliothèques

```python
import math as m
import random as rd
```

<font color='steelblue'> Initialisation des variables: A vous de compléter les 4 premières lignes du code en vous aidant du graphique précédent.

```python
xmin = 0                      # abscisse minimale du rectangle
xmax = 1                      # abscisse maximale du rectangle
ymin = 0                      # ordonées minimale du rectangle
ymax = 0.25                   # ordonées maximale du rectangle

n = 10000                       # Nombre de points tirés au hasard
nD = 0                        # Nombre de points dans le domaine, initialisé à 0 
```

<font color='steelblue'> Création d'une fonction pour calculer image par la fonction $f$:

```python
def image(x):
    y = x*(1-x)*(m.sin(200*x*(1-x)))**2        # calcul de l'image de x par la fonction f en utilisant la fonction sinus via la commande m.sin
    return(y)                                # résultat retourné par la fonction "image"
```

<font color='steelblue'> Il faut ensuite créer tous les points, puis regarder pour chacun d'entre eux, s'il est dans le domaine ou bien à l'extérieur.\
<font color='steelblue'> Dans notre cas, il faut que le point soit sous la courbe pour être dans le domaine.\
<font color='steelblue'> Compléter le code suivant pour atteindre cet objectif, en utilisant la fonction image..

```python
for i in range (0,n):
    x = rd.uniform(xmin, xmax)          # tirage aléatoire (loi uniforme) de l'abscisse du point
    y = rd.uniform(ymin, ymax)          # tirage aléatoire (loi uniforme) de l'ordonnéee du point
    if y <= image(x):
        nD = nD+1
```

<font color='steelblue'> La dernière étape consiste à calculer une valeur approchée de $I$ avec le résultat vu précédemment: $I = Aire(D) \approx \dfrac{n_D}{n} \times {Aire(Rectangle)}$\
<font color='steelblue'> Compléter le code

```python
I = (xmax-xmin)*(ymax-ymin)*nD/n        
print(I)
```

## <font color='steelblue'> Programme complet

```python
import math as m
import random as rd

xmin = 0                      # abscisse minimale du rectangle
xmax = 1                      # abscisse maximale du rectangle
ymin = 0                      # ordonées minimale du rectangle
ymax = 0.25                   # ordonées maximale du rectangle

n = 10000                       # Nombre de points tirés au hasard
nD = 0                        # Nombre de points dans le domaine, initialisé à 0

def image(x):
    y = x*(1-x)*(m.sin(200*x*(1-x)))**2        # calcul de l'image de x par la fonction f
    return(y)
      
for i in range (0,n):
    x = rd.uniform(xmin, xmax)          # tirage de l'abscisse du point
    y = rd.uniform(ymin, ymax)          # tirage de l'ordonnéee du point
    if y <= image(x):
        nD = nD+1

I = (xmax-xmin)*(ymax-ymin)*nD/n        
print(I)   
```

## <font color='steelblue'> Complément n°1: Affichage des points sous la courbe
<font color='steelblue'> Nous allons maintenant compléter le programme précédent pour qu'il affiche les points lorsqu'ils sont positionnés sous la courbe.\

```python
import math as m
import random as rd

xmin = 0                      # abscisse minimale du rectangle
xmax = 1                      # abscisse maximale du rectangle
ymin = 0                      # ordonées minimale du rectangle
ymax = 0.25                   # ordonées maximale du rectangle

n = 10000                     # Nombre de points tirés au hasard
nD = 0                        # Nombre de points dans le domaine, initialisé à 0
X = []                        # Liste des abscisses des points sous la courbe
Y = []                        # Liste des ordonnées des points sous la courbe

def image(x):
    y=x*(1-x)*(m.sin(200*x*(1-x)))**2        # calcul de l'image de x par la fonction f
    return(y)
      
for i in range (0,n):
    x = rd.uniform(xmin, xmax)          # tirage de l'abscisse du point
    y = rd.uniform(ymin, ymax)          # tirage de l'ordonnéee du point
    if y <= image(x):
        nD = nD+1
        X.append(x)                    # ajoute à la liste X la valeur x
        Y.append(y)                    # ajoute à la liste Y la valeur y

plt.figure(figsize=(15,6))             # Choix de la taille du graphique
plt.scatter(X,Y, s=2)                  # création du graphique, s=2 prmet de régler la taille des points
plt.show()                             # affichage du graphique
        
I = (xmax-xmin)*(ymax-ymin)*nD/n        
print(I)
```

## <font color='steelblue'> Complément n°2: Fluctuation de la valeur approchée de $I$
<font color='steelblue'> Nous allons reprendre le programme initial et l'utiliser pour étudier les fluctuations de la valeur aprochée de $I$ obtenue par la méthode de Monte-Carlo.\
<font color='steelblue'> Pour cela nous allons faire tourner notre algorithme plusieurs fois de suite.\
<font color='steelblue'> Il faut donc créer une boucle `for` qui va contenir le programme précédent.    

```python
import math as m
import random as rd
import matplotlib.pyplot as plt  

xmin = 0                      # abscisse minimale du rectangle
xmax = 1                      # abscisse maximale du rectangle
ymin = 0                      # ordonées minimale du rectangle
ymax = 0.25                   # ordonées maximale du rectangle

n = 1000                      # Nombre de points tirés au hasard

k = 10                        # Nombre d'itération de la méthode
ListeI = []                   # Liste qui va contenir les différentes valeurs approchées de I

def image(x):
    y = x*(1-x)*(m.sin(200*x*(1-x)))**2        # calcul de l'image de x par la fonction f
    return(y)

for j in range(0,k):
    nD = 0                        # Nombre de points dans le domaine, initialisé à 0
    for i in range (0,n):
        x = rd.uniform(xmin, xmax)          # tirage de l'abscisse du point
        y = rd.uniform(ymin, ymax)          # tirage de l'ordonnéee du point
        if y <= image(x):
            nD = nD+1
    
    I = (xmax-xmin)*(ymax-ymin)*nD/n
    ListeI.append(I)
print(ListeI)   
```

<font color='steelblue'> Pour visualiser ces fluctuations, nous pouvons créer un nouveau graphique, avec en abscisse le numéro de l'essai et en ordonnée la valeur approchée de $I$ obtenue.

```python
NumeroEssai = [i for i in range(k)]
plt.figure(figsize=(15,6))                         # Choix de la taille du graphique
plt.ylim([0.06,0.10])                              # affiche l'ordonnée entre 0.06 et 0.10
plt.scatter(NumeroEssai,ListeI, s=5)               # création du graphique, s=2 prmet de régler la taille des points
plt.show()                                         # affichage du graphique
```
